## Условие 3: 3 задача


Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости,  с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
import numpy as np
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

Наибольшая прямая зависимость между ценой и жилой площадью (0,7), конструкцией дома 0,7, количеством этажей выше земли (0,6), Кв. метры жилой площади у 15 соседей (0,6).   

```python
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg');
```

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution of price by grade');
```

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('waterfront')
plt.title('Distribution of price by waterfront');
```

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('view')
plt.title('Distribution of price by view');
```

Наибольшая прямая зависимость между ценой и жилой площадью (0,7), конструкцией дома 0,7, количеством этажей выше земли (0,6), Кв. метры жилой площади у 15 соседей (0,6). 
Наиболее равномерно распределена зависимость цены от жилой площади до 7000 кв.м., далее уже единичные объекты с разными ценами, большое влияние на которые вероятно имеют другие факторы.
Зависимость цены от конструкции стремится к равномерному распределению. Свыше оценки 12 конструкций только отдельные объекты. При этом рассматривая распределение каждой оценки конструкции видно, что имеются значительные выбросы в каждой из оценок.
Отсутствие набережной имеет неравномерное распределение с большой дисперсией. Дома с набережной имеют меньший разброс в цене относительно медианы. при этом имеются единичные выбросы.
Вид влияет на цену наравномерно. Достаточно большая дисперсия в распределении цены при рассмотрении влиянии оценки вида. ё
